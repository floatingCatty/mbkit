```python
from mbkit import SquareLattice, LineLattice, ElectronicSpace, EDSolver
from mbkit.operator.models import hubbard
```

```python
lat = SquareLattice(2, 2)
space = ElectronicSpace(lat)
hamiltonian = hubbard(space, t=1.0, U=4.0)
```

```python
solver = EDSolver()

solver.solve(hamiltonian, nsites=4, n_particles=(1,1))
```

```python
solver.energy()
```
