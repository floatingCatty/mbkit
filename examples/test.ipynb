```python
from mbkit import SquareLattice, ElectronicSpace, EDSolver, DMRGSolver
```

```python
lat = SquareLattice(4, 4, max_shell=3)
space = ElectronicSpace(lat)

hamil = space.hubbard(U=4.0, t=1.0)
```

```python
solver = DMRGSolver()
ed_solver = EDSolver()
```

```python
solver.solve(hamil, n_particles=(1,1))
ed_solver.solve(hamil, n_particles=(1,1))
```

```python
solver.energy(), ed_solver.energy()
```
